```python
df1.columns = ['value', 'census_tract', 'avg/std']
```

```python
stats_1 = pd.pivot_table(df1, index=['census_tract'], columns=['avg/std'])
```

```python
stats_1.columns = stats_1.columns.droplevel()
```

```python
stats_1.to_csv('stats_1.csv')
```

```python
with open('stats_1.csv', 'r') as f:
    test = f.readlines()
```

```python
max_dict = dict()
for row in test[1:]:
    tract, avg, std = row.split(',')
    max_dict[tract] = float(avg) + 3*float(std)
```
